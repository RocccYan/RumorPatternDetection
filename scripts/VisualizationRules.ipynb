```python
import pandas as pd
```

```python
rules_stats = pd.read_csv('./20230912-rules.csv', header=0)
```

```python
rules_stats.head()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "roboto" # "futura" #"helvetica" #"cursive" 
# plt.rcParams["font.family"] = "Times New Roman"
# Sample data
A = rules_stats.support.values
B = rules_stats.importance.values
C = rules_stats.coef.values

# Configure the plot
fig, ax = plt.subplots()

# Create the scatter plot
scatter = ax.scatter(A, B, c=C, cmap='coolwarm')

# Add colorbar for the C dimension
cbar = fig.colorbar(scatter)
cbar.outline.set_visible(False)  # Remove colorbar outline
cbar.set_label('Coefficient')

# Set labels and title
ax.set_xlabel('Support')
ax.set_ylabel('Importance')
ax.set_title('Summary of Rules Generated by RuleFit')

# Display the plot
plt.savefig('summary_rules.png', dpi=400)
plt.show()

```
